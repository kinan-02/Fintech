```python
data={"BTC":"/content/BTC.parquet","ETH":"/content/ETH.parquet","GLD":"/content/GLD.parquet","nsdaq" :"/content/nsdaq","SPY":"/content/SPY","UUP" : "/content/UUP.parquet","crude_oil" :"/content/crude_oil","MSTR_STOCK":"/content/MSTR_STOCK"}
```

```python
counter = 0
for name, path in data.items():
      arr = [0]*6
      arr[0] = name
      df = pd.read_parquet(path)
      arr[1] = df['low'].count()
      arr[2] = min(df['low'])
      arr[3] = max(df['high'])
      arr[4] = df["open"].mean()
      arr[5] = df["open"].std()
      dataset.loc[counter] = arr
      counter +=1
```
